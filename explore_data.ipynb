```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_raw_2_min_H1 = pd.read_csv(
    'UK-2-min_data_v1_1254mb/clean/Property_ID=EOH0735.csv')
```

```python

df_raw_half_hour = pd.read_csv('UKDA-9209-csv/csv/eoh_cleaned_half_hourly.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df_raw_half_hour.isnull(), cbar=False, cmap='binary_r', yticklabels=False)

# Customize the plot
plt.title('Missing Data Map (White: Present, Black: Missing)', fontsize=16)
plt.xlabel('Columns', fontsize=12)
plt.ylabel('Rows', fontsize=12)

# Optional: Add column names as x-ticks
plt.xticks(ticks=np.arange(len(df_raw_half_hour.columns)) + 0.5, 
           labels=df_raw_half_hour.columns, rotation=45, ha='right')
plt.tick_params(axis='x', labelsize=10) # Adjust x-axis label font size

plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
